```python
### Forward Connection
class Node:
    def __init__(self, data):
        self.data = data
        ### Current Node has next Node's address
        self.next = None
        
class LinkedListForward:
    def __init__(self):
        ### Here head always stays at 1st Node
        self.head = None
        ### Curr will incrementally updates till the last node
        self.curr = None
        self.takeInputFromList()
    
    def push(self, data):
        newNode = Node(data)
        newNode.next = None
        if self.head == None:
            self.head = newNode
        if self.curr != None:
            self.curr.next = newNode
        self.curr = newNode
        
    def takeInputFromList(self):
        inputList = [int(elem) for elem in input().split()]
        
        for item in inputList:
            self.push(item)
            
    def reverse(self):
        prev = None
        curr = self.head
        while(curr):
            next = curr.next # These actions should take place in order
            curr.next = prev # Do a Dry run on paper for better understanding
            prev = curr
            curr = next
        self.head = prev
    
    def printList(self):
        temp = self.head
        while(temp):
            print (temp.data)
            temp = temp.next
            
    def length(self):
        count = 0
        temp = self.head
        while(temp):
            count += 1
            temp = temp.next
        return count
    
    def insertAtIthIndex(self, i, data):
        if(i < 0 or i > self.length()):
            print('Index out of range')
        elif(i == 0):
            newNode = Node(data)
            newNode.next = self.head
            self.head = newNode
        else:
            count = 0
            curr = self.head
            while(curr):
                count += 1
                if count == i:
                    newNode = Node(data)
                    newNode.next = curr.next
                    curr.next = newNode
                    return
                curr = curr.next ### Incrementing for next Iteration

    def deletingAtIthIndex(self, i):
        if(i < 0 or i >= self.length()):
            print('Index out of range')
        elif(i == 0):
            curr = self.head
            self.head = curr.next
            del curr
        else:
            count = 0
            prev = self.head
            curr = self.head.next # node at index: 1
            while(curr):
                count += 1
                if count == i:
                    prev.next = curr.next
                    del curr
                    return
                prev = prev.next
                curr = curr.next ### Incrementing for next Iteration
```

```python
### Backward Connection
class Node:
    def __init__(self, data):
        self.data = data
        ### Current Node has previous Node's address
        self.prev = None
        
class LinkedListBackWard:
    def __init__(self):
        ### Here head will incrementally updates till the last node
        self.head = None
        self.takeInputFromList()
    
    def push(self, data):
        newNode = Node(data)
        newNode.prev = self.head
        self.head = newNode
        
    def takeInputFromList(self):
        inputList = [int(elem) for elem in input().split()]
        
        for item in inputList:
            self.push(item)
            
    def reverse(self):
        prev = None
        curr = self.head
        while(curr):
            next = curr.prev
            curr.prev = prev
            prev = curr
            curr = next
        self.head = prev

    
    def printList(self):
        temp = self.head
        while(temp):
            print (temp.data)
            temp = temp.prev

    def length(self):
        count = 0
        temp = self.head
        while(temp):
            count += 1
            temp = temp.prev
        return count

        
```
