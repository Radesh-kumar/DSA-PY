```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.head = None
        self.curr = None
        
    def push(self,data):
        newNode = Node(data)
        if(self.head == None):
            self.head = newNode
        if(self.curr != None):
            self.curr.next = newNode
        self.curr = newNode

    def printAll(self):
        curr = self.head
        while(curr):
            print(curr.data)
            curr = curr.next
    def length(self):
        curr = self.head
        count = 0
        while(curr):
            count += 1
            curr = curr.next
        return count

    def reverseIterative(self):
        curr = self.head
        prev = None
        while(curr):
            next = curr.next
            curr.next = prev
            prev = curr
            curr = next
        # Always reset the head after making changes to LL
        self.head = prev
        
    def returnHead(self):
        return self.head
    
    def updateHead(self, node):
        self.head = node
    def hasLoop(self):
        p,q = self.head, self.head
        while True:
            p = p.next
            if(q.next != None):
                q = q.next.next
            else:
                q = None

            if(p == None or q == None):
                return False
            elif(p == q):
                return True

    ### Space Complexity --> O(n)
    def removeLoop(self):
        hs = set()
        prev = self.head
        curr = self.head
        while(curr != None):
            if hex(id(curr)) in hs:
                prev.next = None
                return
            else:
                hs.add(hex(id(curr)))

            prev = curr
            curr = curr.next
        
        
    ### Space Complexity --> O(1)
    def removeLoop2(self):
        slow = fast = head = self.head
        startPointLoop = None
        lastNode = None
        breakPoint = None

        while(slow != None and fast != None and fast.next != None):
            slow = slow.next
            fast = fast.next.next
            if(slow == fast):
                breakPoint = fast
                break
        if(slow == head):
            while(head.next != slow):
                head = head.next
            head.next = None
        elif(slow == fast):
            while (head != breakPoint):
                lastNode = breakPoint
                head = head.next
                breakPoint = breakPoint.next
            startPointLoop = head
            lastNode.next = None
                                    
```
